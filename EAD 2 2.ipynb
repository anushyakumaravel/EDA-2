```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\Users\Lenovo\Desktop\machine l\adult_with_headers.csv")
data
```

```python
# Remove rows with missing values
data.dropna(inplace=True)

# Alternatively, for imputation
data.fillna(data.mode().iloc[0], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Select numerical features
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# Standard Scaling
scaler_standard = StandardScaler()
data[numerical_features] = scaler_standard.fit_transform(data[numerical_features])
data[numerical_features]
```

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
data[numerical_features] = scaler_minmax.fit_transform(data[numerical_features])
data[numerical_features]
```

```python
data.columns = data.columns.str.strip()
```

```python
# Apply One-Hot Encoding to categorical variables with less than 5 categories
categorical_features = ['sex','income']
data = pd.get_dummies(data, columns=categorical_features)
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to categorical variables with more than 5 categories
label_encoder = LabelEncoder()
large_categorical_features = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'native_country']
for feature in large_categorical_features:
    data[feature] = label_encoder.fit_transform(data[feature])
```

```python
# Example new features
data['capital_diff'] = data['capital_gain'] - data['capital_loss']
data['age_hours'] = data['age'] * data['hours_per_week']
data['capital_diff']
data['age_hours']
# Rationale: 'capital_diff' combines two related features, 'age_hours' captures interaction effects between age and working hours.
```

```python
import numpy as np

# Apply log transformation to skewed numerical feature
data['capital_gain'] = np.log1p(data['capital_gain'])
data['capital_gain']
```

```python
from sklearn.ensemble import IsolationForest

# Apply Isolation Forest
iso = IsolationForest(contamination=0.05)
outliers = iso.fit_predict(data[numerical_features])
data = data[outliers == 1]
data
```

```python
# Compare with Correlation Matrix
corr_matrix = data.corr()
print(corr_matrix)
```
